```python
drug_useByAge = pd.read_csv('drugUseByAge.csv', index_col=0, parse_dates=True)
df = pd.DataFrame(drug_useByAge)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
yo = df.iloc[:2,1:5]

labels = 'alcohol use', 'alcohol frequency', 'marijuana use', 'marijuana frequency'
yy = yo.iloc[0:1,]
yyy = yy.values.tolist()
x_list = yyy[0]
```

```python
labels = 'alcohol use', 'alcohol frequency', 'marijuana use', 'marijuana frequency'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(x_list, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
